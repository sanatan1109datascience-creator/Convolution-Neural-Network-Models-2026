```python
import keras.utils
from keras.utils import image_dataset_from_directory
```

## **4.Image Dataset Loading from keras libraries**

```python
train_data=image_dataset_from_directory(
    directory='/content/train',
    labels="inferred",
    label_mode="int",
    batch_size=32,
    image_size=(256, 256),
    verbose=True
)
```

```python
test_data=image_dataset_from_directory(
    directory='/content/test',
    labels="inferred",
    label_mode="int",
    batch_size=32,
    image_size=(256, 256),
    verbose=True
)
```

## **5.Data Normalization**

```python
def process(image,label):

  image=tf.cast(image/255,tf.float32)
  return image,label
```

```python
train_data=train_data.map(process)
test_data=test_data.map(process)
```

```python
train_data
```

## **6.Generating a CNN Network**

```python
## CNN Padding and Pooiling
model=Sequential()
## Padding Operation
model.add(Conv2D(32,kernel_size=9,padding='same',activation='relu',input_shape=(256,256,3)))
## Pooling Operation
model.add(MaxPooling2D(pool_size=4,strides=2,padding='valid'))
model.add(BatchNormalization())

model.add(Conv2D(64,kernel_size=9,padding='same',activation='relu'))
model.add(MaxPooling2D(pool_size=4,strides=2,padding='valid'))
model.add(BatchNormalization())

model.add(Conv2D(128,kernel_size=9,padding='same',activation='relu'))
model.add(MaxPooling2D(pool_size=4,strides=2,padding='valid'))
model.add(BatchNormalization())


## ANN connecting nodes
model.add(Flatten())
model.add(Dense(1000,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(50,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

## **7.Model Compilation**

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=model.fit(train_data,epochs=10,validation_data=test_data)
```
